```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(23)
mu_v1=np.array([0,0,0])
cov_mat1=np.array([[1,0,0],[0,1,0],[0,0,1]])
```

```python
class1_sample=np.random.multivariate_normal(mu_v1,cov_mat1,20)
df=pd.DataFrame(class1_sample,columns=['feature1','feature2','feature3'])
df['target'] = 1
```

```python
df
```

```python
mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)
df1 = pd.DataFrame(class2_sample,columns=['feature1','feature2','feature3'])

df1['target'] = 0
df = df.append(df1,ignore_index=True)
df
```

```python
df=df.sample(40)
```

```python
df
```

```python
import plotly.express as px
fig=px.scatter_3d(df,x=df['feature1'],y=df['feature2'],z=df['feature3'],color=df['target'].astype(str))
# fig.update_traces(marker=dict(size=12,
#                               line=dict(width=2,
#                                         color='DarkSlateGrey')),
#                   selector=dict(mode='markers'))
fig.show()
```

```python
df.iloc[:,0:3]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df.iloc[:,0:3]=scaler.fit_transform(df.iloc[:,0:3])
```

```python
df.iloc[:,0:3]
```

```python
X=df.drop('target',axis=1)
```

```python
df.iloc[:,0:3]=scaler.fit_transform(X)
```

```python
df.iloc[:,0:3]
```

```python
covariance_matrix=np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
```

```python
covariance_matrix
```

```python
eigen_values,eigen_vectors=np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
pc=eigen_vectors[0:2]
pc
```

```python
transformed_df=np.dot(df.iloc[:,0:3],pc.T)
```

```python
new_df=pd.DataFrame(transformed_df,columns=['PC1','PC2'])
new_df['target']=df['target'].values
```

```python
fig=px.scatter(new_df,x='PC1',y='PC2',color=new_df['target'].astype(str))
fig.show()              
               
```
