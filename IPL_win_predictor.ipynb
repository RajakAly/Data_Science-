```python
#It will more easier for analysis if match df and delivery_df 
#is merge.Match_id is the common features on both the datase
#So datasets are merged on Match_id
delivery_df=match_df.merge(delivery_df,on='Match_id')
delivery_df.tail(10)
```

```python
delivery_df = delivery_df[delivery_df['Inning'] == 2]
delivery_df.head()
```

```python
delivery_df.shape
```

```python
delivery_df['Current_runs']=delivery_df.groupby('Match_id').cumsum()['Total_runs_y']
```

```python
delivery_df.head()
```

```python
#Run left to fill the target
delivery_df['Runs_left']=delivery_df['Total_runs_x']-delivery_df['Current_runs']
delivery_df.tail(10)
```

```python
# Suppose 7 balls is played.In the dataset over running is given by Over=2 and ball is represented by Ball =1 
# so remaining ball can be represented as 126-(Over*6 +Ball)
delivery_df['Ball_left']=126-(delivery_df['Over']*6 + delivery_df['Ball'])
delivery_df.head(10)
```

```python
list(delivery_df['Player_dismissed'].unique())
```

```python
delivery_df['Player_dismissed']=delivery_df['Player_dismissed'].fillna("0")
delivery_df['Player_dismissed']=delivery_df['Player_dismissed'].apply(lambda x:x if x=="0" else "1")
delivery_df['Player_dismissed'].dtypes
delivery_df['Player_dismissed'].unique()

```

```python
#Converting the dtypes into int
delivery_df['Player_dismissed']=delivery_df['Player_dismissed'].astype('int64')
delivery_df['Player_dismissed'].dtypes
delivery_df['Player_dismissed'].unique()
```

```python
#Fall of wicket
Wickets=delivery_df.groupby(by=['Match_id']).cumsum()['Player_dismissed'].values
delivery_df['Wickets']=10-Wickets
delivery_df.tail(20)
```

```python
#Run Rate
#Current Run Rate
delivery_df['CRR']=(delivery_df['Current_runs']*6)/(120-delivery_df['Ball_left'])
#Require Run Rate
delivery_df['RRR']=(delivery_df['Runs_left']*6)/delivery_df['Ball_left']
```

```python
delivery_df.head(10)
```

```python
def result_call(df):
    
    return 1 if df['Batting_team'] == df['Winner'] else 0
```

```python
delivery_df['Result'] = delivery_df.apply(result_call,axis=1)
delivery_df.head()
```

```python
final_df=delivery_df[['Batting_team','Bowling_team','City','Runs_left','Ball_left','Wickets','Total_runs_x','CRR','RRR','Result']]
```

```python
final_df=final_df.dropna(axis=0)
final_df.shape
```

```python
final_df=final_df[final_df['Ball_left']!=0]
```

```python
final_df.isnull().sum()
final_df['City'].unique()
```

```python
X=final_df.iloc[:,0:9]
y=final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf1=ColumnTransformer([
    ('trf1',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,2])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
trf2=LogisticRegression(solver='liblinear')
```

```python
from sklearn.pipeline import Pipeline
pipe=Pipeline([
    ('trf1',trf1),
    ('trf2',trf2)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_hat=pipe.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_hat)
```

```python
pipe.predict_proba(X_test)
```
